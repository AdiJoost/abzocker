```python
df_apple = df[df.Symbol == "AAPL"]
```

```python
#fig, ax = plt.subplots(2,3)
plt.title("APPLE")
plt.plot(df_apple["Date"], df_apple["Close"])
plt.gca().xaxis.set_major_locator(MultipleLocator(100))
plt.xticks(rotation=90)
plt.tight_layout()


#plt.plot(df_apple["Date"], df_apple["Open"])
plt.legend()
plt.show()
```

```python
start_date = "2010-01-01"
end_date = "2010-12-30"
df_apple_range = df_apple[(df_apple['Date'] >= start_date) & (df_apple['Date'] <= end_date)]
plt.title("APPLE")
plt.plot(df_apple_range["Date"], df_apple_range["Close"])
plt.gca().xaxis.set_major_locator(MultipleLocator(30))
plt.xticks(rotation=90)
plt.tight_layout()


#plt.plot(df_apple["Date"], df_apple["Open"])
plt.show()
```
